```python
import os

def menu():

  print("*" * 40)
  print("TO DO LIST".center(40))
  print("*" * 40)
  print("1.VIEW TASKS".rjust(20))
  print("2.ADD TASK".rjust(18))
  print("3.MARK AS DONE".rjust(22))
  print("4.DELETE TASK".rjust(21))
  print("5.UPDATE TASK".rjust(21))
  print("6.Exit".rjust(14))

def view_task():
  with open("tasks.txt","r") as file:
    tasks = file.readlines()
    status=''
    if tasks:
        for index, task in enumerate(tasks,1):
            if "[x]" in task:
                status = "Done"
            else:
                status = " Not Done"
            task_text = task.replace("[ ]", "").replace("[x]", "").strip()
            print(f"{index} {task_text} - {status}")
    else:
        print("TO DO LIST EMPTY....:(")

def add_task():
    task = input("Enter Task To Be Added : ")
    with open("tasks.txt", "a") as file:
        file.write(f'{task} [ ]\n')
    print("TASK ADDED SUCCESSFULLY....:)")

def mark_as_done():
    view_task()
    try:
        task_num = int(input("Enter the task number to mark as done: "))
        with open("tasks.txt", "r") as file:
            tasks = file.readlines()
        if 1 <= task_num <= len(tasks):
            task = tasks[task_num - 1]
            if "[ ]" in task:
                tasks[task_num - 1] = task.replace("[ ]", "[x]")
                with open("tasks.txt", "w") as file:
                    file.writelines(tasks)
                print("Task Marked As Done")
            else:
                print("Task Is Already Marked As Done")
        else:
            print("INVALID Task Number")
    except ValueError:
        print("INVALID INPUT!!!...Please Enter A Number....:( ")

def delete():
    view_task()
    try:
        task_num = int(input("Enter Task Number To Be Deleted : "))
        with open("tasks.txt", "r+") as file:
            tasks = file.readlines()
        if 1 <= task_num < len(tasks):
            deleted_task = tasks.pop(task_num - 1)
            with open("tasks.txt", "w") as file:
                file.writelines(tasks)
            print(f"Task deleted: {deleted_task.strip()}")
        else:
            print("Invalid Task Number")

    except ValueError:
        print("INVALID INPUT!!!...Please Enter A Number....:( ")

def update():
    view_task()
    try:
        task_index = int(input("Enter Task Number To Be Updated : "))
        with open("tasks.txt", "r") as file:
            tasks = file.readlines()
        if 1 <= task_index <= len(tasks):
            new_task = input("Enter The Updated Task : ")
            tasks[task_index - 1] = f"{new_task} [ ]\n"
            with open("tasks.txt", "w") as file:
                file.writelines(tasks)
            print("Task Updated Successfully.....:)")
        else:
            print("INVALID TASK NUMBER....:(")
    except ValueError:
        print("INVALID INPUT Please Enter A Number ")
def main():

    while True:
        menu()
        try:
          choice = int(input("Enter Your Choice: "))
          if choice == 1:
              view_task()
          elif choice == 2:
              add_task()
          elif choice == 3:
              mark_as_done()
          elif choice == 4:
              delete()
          elif choice == 5:
              update()
          elif choice == 6:
              print("Exiting TO-DO List.........:(")
              break
          else:
              print("INVALID CHOICE.... Please Enter A Number Between 1 TO 6....")
        except:
          print('INVALID INPUT....Enter A Number....:(')

if __name__ == "__main__":
    if not os.path.exists("tasks.txt"):
        with open("tasks.txt", "w"):
            pass
    main()
```
