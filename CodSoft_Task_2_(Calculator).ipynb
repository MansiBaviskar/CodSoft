```python
def menu():

  print('*' * 40)
  print('__CALCULATOR__'.center(40))
  print('*' * 40)
  print('OPERATIONS'.rjust(18))
  print('1.ADDITION'.rjust(20))
  print('2.SUBTRACTION'.rjust(23))
  print('3.MULTIPLICATION'.rjust(26))
  print('4.DIVISION'.rjust(20))
  print('5.FLOOR DIVISION'.rjust(26))
  print('6.MODULUS'.rjust(19))
  print('7.EXPONENTIATION'.rjust(26))
  print('8.EXIT'.rjust(16))

def add(a,b):
    return a+b

def sub(a,b):
    return a - b

def mul(a,b):
    return a * b

def div(a,b):
    return a / b

def f_div(a,b):
    return a // b

def mod(a,b):
    return a % b

def pow(a,b):
    return a ** b

def main():

  while True:
      menu()

      try:
        choice = int(input('ENTER OPERATION NUMBER :'))

      except:
          print('Enter Valid Input.....:(')
          continue

      if choice == 8:
          print("Exiting the calculator........... Goodbye!!!!!!!")
          break

      if choice in (1,2,3,4,5,6,7):

        try:

          in1 = float(input('ENTER NUMBER A :'))
          in2 = float(input('ENTER NUMBER B :'))

        except:

          print('Enter Valid Input.....:(')
          continue

        if choice == 1:
            print('Result of {} + {} = {}'.format(in1,in2,add(in1,in2)))

        elif choice == 2:
            print('Result of {} - {} = {}'.format(in1,in2,sub(in1,in2)))

        elif choice == 3:
            print('Result of {} * {} = {}'.format(in1,in2,mul(in1,in2)))

        elif choice == 4:
            print('Result of {} / {} = {}'.format(in1,in2,div(in1,in2)))

        elif choice == 5:
            print('Result of {} // {} = {}'.format(in1,in2,f_div(in1,in2)))

        elif choice == 6:
            print('Result of {} % {} = {}'.format(in1,in2,mod(in1,in2)))

        elif choice == 7:
            print('Result of {} ** {} = {}'.format(in1,in2,pow(in1,in2)))

      else:
          print("INVALID CHOICE.... Please Enter A Number Between 1 TO 7....")

if __name__ == "__main__":
    main()
```
