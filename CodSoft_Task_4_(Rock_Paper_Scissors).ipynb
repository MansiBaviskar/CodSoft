```python
import random

ROCK = "1"
PAPER = "2"
SCISSORS = "3"

def display_options():
    print("\nLet's play Rock, Paper, Scissors!")
    print("Your choices:")
    print(f"{ROCK} - Rock")
    print(f"{PAPER} - Paper")
    print(f"{SCISSORS} - Scissors")

def get_user_choice():
    while True:
        try:
            user_choice = int(input("Enter your choice (1 for Rock, 2 for Paper, or 3 for Scissors): "))
            if user_choice in [1, 2, 3]:
                return str(user_choice)
                break
            else:
                print("Invalid choice. Please enter 1 for Rock, 2 for Paper, or 3 for Scissors.")
        except ValueError:
            print("Invalid input. Please enter a number (1, 2, or 3).")

def get_computer_choice():
    return str(random.randint(1, 3))

def announce_round_result(user_choice, comp_choice, user_score, comp_score):
    choices_mapping = {ROCK: "Rock", PAPER: "Paper", SCISSORS: "Scissors"}
    user_choice_str = choices_mapping[user_choice]
    comp_choice_str = choices_mapping[comp_choice]

    print(f"\nYou chose {user_choice_str}.")
    print(f"The computer chose {comp_choice_str}.")

    if user_choice == comp_choice:
        print("It's a tie!")
    elif (user_choice == ROCK and comp_choice == SCISSORS) or \
         (user_choice == PAPER and comp_choice == ROCK) or \
         (user_choice == SCISSORS and comp_choice == PAPER):
        print("You win this round!")
        user_score += 1
    else:
        print("Computer wins this round!")
        comp_score += 1

    print(f"\nYour Score: {user_score} | Computer Score: {comp_score}")
    return user_score, comp_score

def announce_final_result(user_score, comp_score):
    print("\nGame Over!")
    print(f"Your Final Score: {user_score} | Computer Final Score: {comp_score}")

    if user_score > comp_score:
        print("Congratulations! You win!")
    elif user_score < comp_score:
        print("Sorry, computer wins. Better luck next time.")
    else:
        print("It's a tie!")

if __name__ == "__main__":
    display_options()
    while True:
      try:
          rounds = int(input("\nHow many rounds do you want to play? "))
          if rounds <= 0:
              print("Please enter a positive integer for the number of rounds.")
              exit()
          else:
            break
      except ValueError:
          print("Invalid input. Please enter a valid integer.")
          exit()


    user_score = 0
    comp_score = 0

    for round_num in range(1, rounds + 1):
        print(f"\nRound {round_num}:")
        user_choice = get_user_choice()
        comp_choice = get_computer_choice()
        user_score, comp_score = announce_round_result(user_choice, comp_choice, user_score, comp_score)

    announce_final_result(user_score, comp_score)
```
